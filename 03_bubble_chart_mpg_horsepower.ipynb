```python
import pandas as pd 
import plotly.graph_objs as go 
import plotly.offline as pyo 
```

```python
df = pd.read_csv(r'Data\mpg.csv')
```

```python
df.head()
```

```python
data = [go.Scatter(
                    x=df['horsepower'], y=df['acceleration'],text=df['mpg'],
                    mode='markers',
                    marker=dict(size=df['displacement']/5, color=df['cylinders'], showscale=True)
                   )
        ]
```

```python
layout = go.Layout(title='Car performance', xaxis=dict(title='horsepower'), yaxis=dict(title='Acceleration'), hovermode='closest' )
```

```python
fig = go.Figure(data=data, layout=layout)
```

```python
pyo.plot(fig, filename='bubble_chart_mpg_horsepower.html')
```
